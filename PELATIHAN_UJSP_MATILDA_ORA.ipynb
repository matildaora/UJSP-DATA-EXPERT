```python

df = pd.read_csv('Mall_Customers.csv')
df.head(3)
```

```python
#Mengubah Nama Kolom

df = df.rename(columns={'Gender': 'gender',
                        'Age': 'age',
                        'Annual Income (k$)': 'annual_income',
                        'Spending Score (1-100)': 'spending_score'})

#Mengubah Data kategorik menjadi numerik
df['gender'].replace(['Female', 'Male'], [0,1], inplace=True)
df.head()
```

```python
#Delete Kolom Customers ID dan Gender

X = df.drop(['CustomerID', 'gender'], axis=1)
```

```python
#Import Algoritma

from sklearn.cluster import KMeans
```

```python
#Membuat List dari Nilai K 1-11 (inertia_)
from sklearn.cluster import KMeans

clusters = []
for i in range(1, 11):
    km = KMeans(n_clusters=i).fit(X)
    clusters.append(km.inertia_)

print(clusters)
```

```python
#import library untuk visualisasi

import matplotlib.pyplot as plot
%matplotlib inline
import seaborn as sns
```

```python
#membuat plot inertia

fig, ax = plt.subplots(figsize=(8,4))
sns.lineplot(x=list(range(1, 11)), y=clusters, ax=ax)
ax.set_title('Cari Elbow')
ax.set_xlabel('Clusters')
ax.set_ylabel('Inertia')
```

```python
#Membuat Objek KMeans

km5 = KMeans(n_clusters=5).fit(X)

#Menambahkan Kolom Label pada Dataset

X['Labels'] = km5.labels_
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

# Contoh DataFrame X
import pandas as pd
X = pd.DataFrame({
    'annual_income': [15, 16, 17, 18, 19, 20],
    'spending_score': [39, 81, 6, 77, 40, 76]
})

# KMeans Clustering dengan 5 cluster
kmeans = KMeans(n_clusters=5, random_state=42)
X['Labels'] = kmeans.fit_predict(X)  # Menambahkan label ke X

# Plot hasil clustering
plt.figure(figsize=(8, 4))
sns.scatterplot(x=X['annual_income'], y=X['spending_score'], hue=X['Labels'].astype(str),
                palette=sns.color_palette('hls', 5))
plt.title('KMeans dengan 5 Cluster')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
import pandas as pd

def get_category_dict(category_file):
    # Membaca file JSON ke DataFrame
    category = pd.read_json(category_file, orient="records")
    category = pd.DataFrame(category["items"].values.tolist())

    # Mengembalikan dictionary dengan id sebagai key dan title sebagai value
    return {
        cat.id: cat.snippet.get("title")
        for cat in category.itertuples(index=False)
    }

# Memanggil fungsi untuk membaca kategori
try:
    category_dict = get_category_dict("category.json")
    print(category_dict)  # Menampilkan hasil
except FileNotFoundError:
    print("File category.json tidak ditemukan!")
```
